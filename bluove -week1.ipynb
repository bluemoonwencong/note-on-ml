```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(range(100))
add = 2*np.random.randn(100)
y = 0.34*x + 10 + add
```

```python
plt.figure()
plt.scatter(x, y, color='black')
plt.plot(x, 0.34*x+10, color='green', label='y=0.34*x+10')
p0=plt.show()
```

```python
# 最小二乘法
# y = w1*x + w0
# cost = np.mean((w1*x+w0 - y)**2)
w1 = ( np.dot(y, x) - np.mean(y)*np.sum(x) ) / (np.dot(x,x) - np.mean(x)*np.sum(x))
w0 = np.mean(y-w1*x)
cost = np.mean((w1*x+w0 - y)**2)
print(w0, w1,cost)
plt.figure()
plt.scatter(x, y, color='black')
plt.plot(x, 0.34*x+10, color='red', label='y=0.34*x+10')
plt.plot(x, w1*x+w0, color='green', label='y=w1*x+w0')
plt.show()
```

```python
# 岭回归, L2正则化
# y = w1*x + w0
# cost = np.mean((w1*x+w0 - y)**2) + 0.5*alpha*(w1**2+w0**2)
alpha=0.25
w1 = ( np.dot(y, x) - (1/(1+alpha))*np.mean(y)*np.sum(x) ) / (np.dot(x,x) + x.__len__()*alpha - (1/(1+alpha))*np.mean(x)*np.sum(x))
w0 = np.mean(y-w1*x)
cost = np.mean((w1*x+w0 - y)**2) + 0.5*alpha*(w1**2+w0**2)
print(w0, w1,cost)
plt.figure()
plt.scatter(x, y, color='black')
plt.plot(x, 0.34*x+10, color='red', label='y=0.34*x+10')
plt.plot(x, w1*x+w0, color='green', label='y=w1*x+w0')
plt.show()
```

```python
# Lasso, L1正则化，不知道怎么写，只有调包了。
xx = np.zeros(100).reshape(100,1)
for i in range(100): xx[i] = np.array(list([x[i] ]))

from sklearn import linear_model
clf = linear_model.Lasso(alpha = 0.25)
clf.fit(xx,y)

# cost = np.mean((w1*x+w0 - y)**2) + 0.5*alpha*(np.abs(w1)+np.abs(w0))
# print(w0, w1,cost)
plt.figure()
plt.scatter(x, y, color='black')
plt.plot(x, 0.34*x+10, color='red', label='y=0.34*x+10')
plt.plot(x, clf.predict(xx), color='green', label='y=w1*x+w0')
plt.show()
# 同样都是alpha=0.25，调包的效果格外好一些。
```
